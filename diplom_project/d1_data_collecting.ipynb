# <center> Скопировала таблицу безработных с сайта "gogov.ru" путем парсинга

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Create an URL object
url = 'https://gogov.ru/articles/unemployment-rate'
# Create object page
page = requests.get(url,headers = {"User-Agent":"Mozilla/5.0"})

print(page.status_code)
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# Obtain information from tag <table>
table1 = soup.find('table', id='m-table')
table1
```

```python
# Obtain every title of columns with tag <th>
headers = []
for i in table1.find_all('th'):
    title = i.text
    headers.append(title)
```

```python
# Create a dataframe
mydata = pd.DataFrame(columns = headers)
```

```python
# Create a for loop to fill mydata
for j in table1.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(mydata)
    mydata.loc[length] = row
```

```python
mydata
```

```python
# Export to csv
mydata.to_csv('social_russia_data/unemployed.csv', index=False)

```
